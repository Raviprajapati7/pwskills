Q2. Given a Pandas DataFrame df with columns 'A', 'B', and 'C', write a Python function to re-index the DataFrame with a new index that starts from 1 and increments by 2 for each row

Answer2:

```python
def reindex_dataframe(df):
    new_index = pd.RangeIndex(start=1, step=2, stop=len(df) * 2)
    df_reindexed = df.reset_index(drop=True)
    df_reindexed.index = new_index
    return df_reindexed

# Example usage
df1 = pd.DataFrame({'A': [10, 20, 30, 40, 50],
                   'B': [100, 200, 300, 400, 500],
                   'C': [1000, 2000, 3000, 4000, 5000]})

reindexed_df = reindex_dataframe(df1)
print(reindexed_df)
```

Q3. You have a Pandas DataFrame df with a column named 'Values'. Write a Python function that iterates over the DataFrame and calculates the sum of the first three values in the 'Values' column. The function should print the sum to the console.

```python
def calculate_sum(df):
    # Initialize a variable to store the sum
    total_sum = 0
    
    # Iterate over the DataFrame using iterrows()
    for index, row in df.iterrows():
        # Add the value to the total sum
        total_sum += row['Values']
        
        # Break the loop after calculating the sum of the first three values
        if index == 2:
            break
    
    # Print the sum to the console
    print("Sum of the first three values:", total_sum)

# Example DataFrame
df = pd.DataFrame({'Values': [10, 20, 30, 40, 50]})

# Calculate the sum of the first three values
calculate_sum(df)
```

Q4. Given a Pandas DataFrame df with a column 'Text', write a Python function to create a new column 'Word_Count' that contains the number of words in each row of the 'Text' column.

```python
import pandas as pd

def add_word_count(df):
    # Apply a lambda function to count the number of words in each row of the 'Text' column
    df['Word_Count'] = df['Text'].apply(lambda x: len(str(x).split()))
    
    return df
# Example DataFrame
df = pd.DataFrame({'Text': ['Hello, how are you?',
                            'I am doing great.',
                            'Python is awesome!']})

# Add the 'Word_Count' column
df = add_word_count(df)

# Print the updated DataFrame
print(df)
```

Q5. How are DataFrame.size() and DataFrame.shape() different?

Answer 5: The DataFrame.size and DataFrame.shape are both attributes of a pandas DataFrame, but they provide different information about the DataFrame:

1 DataFrame.size: This attribute returns the total number of elements in the DataFrame, which is calculated as the product of the number of rows and the number of columns. It represents the total size of the DataFrame in terms of elements.

2 DataFrame.shape: This attribute returns a tuple representing the dimensions of the DataFrame. The tuple contains two values: the number of rows and the number of columns, respectively. It provides information about the structure of the DataFrame.

Here's a summary of the differences between DataFrame.size and DataFrame.shape:

1 DataFrame.size gives the total number of elements in the DataFrame.

2 DataFrame.shape gives the dimensions of the DataFrame as a tuple (number of rows, number of columns).

Here's an example to illustrate the difference:

```python
# Example DataFrame
df = pd.DataFrame({'A': [1, 2, 3],
                   'B': [4, 5, 6],
                   'C': [7, 8, 9]})

# DataFrame.size
size = df.size

# DataFrame.shape
shape = df.shape

print("DataFrame size:", size)
print("DataFrame shape:", shape)
```

Q8. You have a Pandas DataFrame df with columns 'A', 'B', and 'C'. Write a Python function that selects all rows where the value in column 'A' is greater than 5 and the value in column 'B' is less than 10. The function should return a new DataFrame that contains only the selected rows.

For example, if df contains the following values:

A B C

0 3 5 1

1 8 2 7

2 6 9 4

3 2 3 5

4 9 1 2


```python
def select_rows(df):
    # Select rows where 'A' > 5 and 'B' < 10
    selected_rows = df[(df['A'] > 5) & (df['B'] < 10)]
    
    return selected_rows

# Example DataFrame
df = pd.DataFrame({'A': [3, 8, 6, 2, 9],
                   'B': [5, 2, 9, 3, 1],
                   'C': [1, 7, 4, 5, 2]})

# Select rows based on conditions
selected_df = select_rows(df)

# Print the selected DataFrame
print(selected_df)
```

Q9. Given a Pandas DataFrame df with a column 'Values', write a Python function to calculate the mean,median, and standard deviation of the values in the 'Values' column.

```python
import pandas as pd

def calculate_stats(df):
    # Calculate mean
    mean_value = df['Values'].mean()
    
    # Calculate median
    median_value = df['Values'].median()
    
    # Calculate standard deviation
    std_value = df['Values'].std()
    
    # Create a dictionary to store the results
    stats = {
        'Mean': mean_value,
        'Median': median_value,
        'Standard Deviation': std_value
    }
    
    return stats
# Example DataFrame
df = pd.DataFrame({'Values': [1, 2, 3, 4, 5]})

# Calculate statistics
stats = calculate_stats(df)

# Print the results
for stat, value in stats.items():
    print(f"{stat}: {value}")
```

Q10. Given a Pandas DataFrame df with a column 'Sales' and a column 'Date', write a Python function to create a new column 'MovingAverage' that contains the moving average of the sales for the past 7 days for each row in the DataFrame. The moving average should be calculated using a window of size 7 and should include the current day.

Answer10:

```python
def calculate_moving_average(df):
    # Sort the DataFrame by the 'Date' column
    df = df.sort_values('Date')

    # Calculate the moving average using a window of 7 days
    df['MovingAverage'] = df['Sales'].rolling(window=7, min_periods=1).mean()

    return df

# Example DataFrame
df = pd.DataFrame({
    'Date': pd.date_range(start='2023-01-01', periods=10),
    'Sales': [10, 15, 12, 8, 20, 18, 25, 14, 9, 16]
})

# Calculate the moving average
df = calculate_moving_average(df)

# Print the updated DataFrame
print(df)
```

Q11. You have a Pandas DataFrame df with a column 'Date'. Write a Python function that creates a new column 'Weekday' in the DataFrame. The 'Weekday' column should contain the weekday name (e.g. Monday, Tuesday) corresponding to each date in the 'Date' column. For example, if df contains the following values:

Date

0 2023-01-01

1 2023-01-02

2 2023-01-03

3 2023-01-04

4 2023-01-05

Your function should create the following DataFrame:

Date Weekday

0 2023-01-01 Sunday

1 2023-01-02 Monday

2 2023-01-03 Tuesday

3 2023-01-04 Wednesday

4 2023-01-05 Thursday

The function should return the modified DataFrame.

Answer11:

```python
def add_weekday_column(df):
    # Convert the 'Date' column to datetime type if necessary
    df['Date'] = pd.to_datetime(df['Date'])

    # Extract the weekday name and assign it to the 'Weekday' column
    df['Weekday'] = df['Date'].dt.day_name()

    return df
# Example DataFrame
df = pd.DataFrame({'Date': ['2023-01-01', '2023-01-02', '2023-01-03', '2023-01-04', '2023-01-05']})

# Add the 'Weekday' column
df = add_weekday_column(df)

# Print the modified DataFrame
print(df)
```

Q12. Given a Pandas DataFrame df with a column 'Date' that contains timestamps, write a Python function to select all rows where the date is between '2023-01-01' and '2023-01-31'.

Answer12:

```python
def select_rows_between_dates(df):
    # Convert 'Date' column to datetime type if necessary
    df['Date'] = pd.to_datetime(df['Date'])
    
    # Define the start and end dates
    start_date = pd.to_datetime('2023-01-01')
    end_date = pd.to_datetime('2023-01-31')
    
    # Select rows where the date is between the start and end dates
    selected_rows = df[(df['Date'] >= start_date) & (df['Date'] <= end_date)]
    
    return selected_rows
# Example DataFrame
df = pd.DataFrame({'Date': ['2023-01-01', '2023-01-15', '2023-02-01', '2023-01-10']})

# Select rows between '2023-01-01' and '2023-01-31'
selected_df = select_rows_between_dates(df)

# Print the selected DataFrame
print(selected_df)
```
